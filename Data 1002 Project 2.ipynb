```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#converting the data into a dataframe
df = pd.read_csv('/Users/agnelvarghesepaikkatt/Downloads/SDG_goal3_clean.csv')

#making a dataframe for specific years
df2 = df.loc[df['Year'] == 2000]
df3 = df.loc[df['Year'] == 2005]
df4 = df.loc[df['Year'] == 2010]
df5 = df.loc[df['Year'] == 2015]

#grouping the dataset by region
df1 = df.groupby(['Region'])

#Takiing the sum of the column maternal mortality ratio
maternal_mortality = df1["Maternal mortality ratio"].sum()
dfy = pd.DataFrame(maternal_mortality)
display(dfy.set_axis(["Maternal mortality"],axis=1))

#Taking the sum the column nurse?midwife density 
nursemidwife_density = df1["Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"].sum()
dfyy = pd.DataFrame(nursemidwife_density)
display(dfyy.set_axis(["Nurse availability"],axis=1))

#producing the first scatter plot without removing outliers or arranging the charts by year
sns.scatterplot(data=df,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')


plt.xlabel('nurse or midwife density\n')
plt.ylabel('maternal mortlity ratio\n')

plt.title('Does density of healthcare workers namely nurses or midwives in a region affect maternal mortality\n')

plt.legend()

plt.show()

#Scatterplots with outliers removed
graph = sns.scatterplot(data=df,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')
graph.set_ylim(0,150)

plt.xlabel('nurse or midwife density without outliers for clarity\n')
plt.ylabel('maternal mortlity ratio\n')

plt.title('Does density of healthcare workers namely nurses or midwives in a region affect maternal mortality WITHOUT OUTLIERS\n')

plt.legend()

plt.show()

#plotting subplots
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(20,10))
fig.suptitle('Does density of healthcare workers namely nurses or midwives in a region affect maternal mortality')

graph = sns.scatterplot(ax=axes[0, 0],data=df2,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')
graph.set_ylim(0,150)
axes[0, 0].set_title(2000)

graph = sns.scatterplot(ax=axes[0, 1],data=df3,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')
graph.set_ylim(0,150)
axes[0, 1].set_title(2005)


graph = sns.scatterplot(ax=axes[1, 0],data=df4,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')
graph.set_ylim(0,150)
axes[1, 0].set_title(2010)


graph = sns.scatterplot(ax=axes[1, 1],data=df5,x = "Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE"  ,y = "Maternal mortality ratio"  , hue='Region')
graph.set_ylim(0,150)
axes[1, 1].set_title(2015)



```
